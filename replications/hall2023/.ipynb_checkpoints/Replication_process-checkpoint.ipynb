```python
import pandas as pd
import numpy as np
from scipy import stats

import statsmodels.formula.api as smf

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Bring in the main analysis dataset
full = pd.read_stata("primary_analysis.dta")
full
```

```python
full.info(verbose=True)
```

```python
#figure 1

# read in data
#We already read the data above: 'full'

# run linear regression
reg = smf.ols('dwnom1 ~ winner_hall_snyder_score', data=full).fit()

# create scatter plot with regression line
plt.figure()
plt.scatter(x=full['winner_hall_snyder_score'], y=full['dwnom1'], color=np.where(full['dem']==1, 'dodgerblue', 'red'))
plt.plot(full['winner_hall_snyder_score'], reg.predict(full), color='black')
plt.xlabel('Contribution-based Estimated Ideology')
plt.ylabel('Observed DW-NOMINATE Score')
plt.tick_params(axis='y', direction='in', labelleft=True, labelright=False)
plt.savefig('compare.pdf')
plt.show()

# overall correlation
print(np.corrcoef(full['winner_hall_snyder_score'], full['dwnom1'])[0,1])

# party-specific correlations
print(np.corrcoef(full.loc[full['dem']==1, 'winner_hall_snyder_score'], full.loc[full['dem']==1, 'dwnom1'])[0,1])
print(np.corrcoef(full.loc[full['dem']==0, 'winner_hall_snyder_score'], full.loc[full['dem']==0, 'dwnom1'])[0,1])
```

```python
#figure 2


# Subset data to only elections within .2 margin and above median distance
rd_data = full[(full['margin'] <= 0.2) & (full['absdist'] > 0.1095)]

# Make the binned averages, first the bins to the left
count = 0
bin_size = 0.02
binx_left = np.zeros(len(np.arange(-0.2, 0, bin_size)))
biny_left = np.zeros(len(binx_left))
last = -0.2
for j in np.arange(-0.2, 0, bin_size):
    biny_left[count] = np.mean(rd_data['dv'][(rd_data['rv'] >= j-bin_size) & (rd_data['rv'] < j)])
    binx_left[count] = (j+last)/2
    last = j
    count += 1

# Now the bins on the right
count = 0
bin_size = 0.02
binx_right = np.zeros(len(np.arange(0.02, 0.2, bin_size)))
biny_right = np.zeros(len(binx_right))
last = 0
for j in np.arange(0.02, 0.2, bin_size):
    biny_right[count] = np.mean(rd_data['dv'][(rd_data['rv'] >= j-bin_size) & (rd_data['rv'] < j)])
    binx_right[count] = (j+last)/2
    last = j
    count += 1

# Fit linear regression models for the left and right sides
reg1 = np.polyfit(rd_data['rv'][rd_data['rv'] < 0], rd_data['dv'][rd_data['rv'] < 0], 1)
reg2 = np.polyfit(rd_data['rv'][rd_data['rv'] > 0], rd_data['dv'][rd_data['rv'] > 0], 1)

fits1 = reg1[1] + reg1[0] * rd_data['rv'][rd_data['rv'] < 0]
fits2 = reg2[1] + reg2[0] * rd_data['rv'][rd_data['rv'] > 0]

sizes = 1

plt.figure()
#Had to change gray50 -> gray to fit the matplotlib format
plt.plot(rd_data['rv'], rd_data['dv'], 'o', color='gray', markersize=3)
plt.plot(binx_left, biny_left, 'o', color='black', markersize=8)
plt.plot(binx_right, biny_right, 'o', color='black', markersize=8)
plt.plot(rd_data['rv'][rd_data['rv'] < 0], fits1, color='black', linewidth=4)
plt.plot(rd_data['rv'][rd_data['rv'] > 0], fits2, color='black', linewidth=4)
#Had to change gray10 -> lightgray to fit the matplotlib format
plt.axvline(x=0, color='lightgray', linestyle='--', linewidth=2)
plt.ylim(0.35, 0.8)
plt.xlim(-0.2, 0.2)
plt.xlabel('Extreme Candidate Primary Election Winning Margin', fontsize=14)
plt.ylabel('General Election Vote Share', fontsize=14)
plt.xticks(np.arange(-0.2, 0.3, 0.1), fontsize=12)
plt.yticks(np.arange(0.4, 0.9, 0.2), fontsize=12)
```
